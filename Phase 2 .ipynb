##### 1.3: Handle Missing Values
Ensures no empty rows remain.
Missing categorical values are replaced with "Unknown".

```python
# Count missing values per column
df.isnull().sum()
```

```python
# Drop rows that are completely empty
df = df.dropna(how='all')
```

```python
# Fill remaining missing values with 'Unknown'
df = df.fillna("Unknown")
```

##### 1.4: Check and Remove Duplicates
Prevents repeated rows from skewing analysis.

```python
# Count duplicates
df.duplicated().sum()
```

##### Step1.5: Rename Columns (for clarity)
This Makes the dataset easier to interpret.

```python
df.rename(columns={ "Column1": "Style", "Column2": "Price", "Column3": "Rating", "Column4": "Size", "Column5": "Season", "Column6": "NeckLine", "Column7": "SleeveLength", "Column8": "WaistLine", "Column9": "Material", "Column10": "FabricType", "Column11": "Decoration", "Column12": "PatternType", "Column13": "Class" }, inplace=True)
df.head()
```

##### 1.6: Standardize Categories
Fixes inconsistent capitalization

```python
# Convert text columns to consistent format
df['Style'] = df['Style'].str.capitalize()
df['Price'] = df['Price'].str.lower()
df['Season'] = df['Season'].str.capitalize()
```

##### 1.7: Convert Data Types
Ensures ratings are treated as numbers for analysis.

```python
# Convert Rating to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```
